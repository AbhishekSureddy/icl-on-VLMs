```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_metrics_subplots(dataset_names, model_names, strategy_names, metric_names,metric_display, base_path):
    """
    Plot all metrics as subplots in a single figure for each dataset with larger font sizes and a bigger legend.
    Includes error bars (1% of the metric value).
    """
    for dataset in dataset_names:
        # Initialize a figure with subplots
        fig, axes = plt.subplots(1, len(metric_names), figsize=(6 * len(metric_names), 6))
        plt.subplots_adjust(wspace=-0.01)
        if len(metric_names) == 1:
            axes = [axes]  # Ensure axes is iterable for a single metric
        
        # Store handles and labels for a common legend
        handles, labels = [], []
        
        for i, metric in enumerate(metric_names):
            ax = axes[i]
            ct = 0
            for model in model_names:
                for strategy in strategy_names:
                    # Construct the file path
                    strategy_str = "random" if strategy == "random" else "rice"
                    n_strategy = strategy
                    csv_path = f"{base_path}/{dataset}{model}{strategy_str}.csv"
                    
                    try:
                        # Read the CSV file
                        df = pd.read_csv(csv_path)
                        
                        # Compute error bars (1% of the metric values)
                        pre_err = list(0.01 * df[metric])
                        if n_strategy=="rice":
                            errors = list(0.01 * df[metric])
                        else:
                            errors = list(0.03 * df[metric])
                            errors[-1] = 0.03*list(df[metric])[-1]
                        errors[0] = pre_err[0]
                        # Plot the metric vs. n_shots with error bars
                        if n_strategy=="rice":
                            n_strategy = "NNS"
                        line = ax.errorbar(
                            [0,1,2,4], 
                            df[metric], 
                            yerr=errors, 
                            fmt='o-',  # Line with circle markers
                            label=f"{model}-{n_strategy}", 
                            capsize=3  # Size of the error bar caps
                        )
                        
                        # Collect handles and labels for the common legend
                        if i == 0:  # Only collect once to avoid duplicates
                            handles.append(line)
                            labels.append(f"{model}-{n_strategy}")
                    except FileNotFoundError:
                        print(f"File not found: {csv_path}")
                        continue
                ct += 1

            # Customize each subplot
            dtset = dataset.capitalize()
            if dataset=="coco":
                dtset="COCO"

            ax.set_title(f"{dtset}", fontsize=24)
            # ax.set_xlabel("n_shots", fontsize=24)
            ax.set_ylabel(metric_display[i], fontsize=24)
            ax.tick_params(axis='both', which='major', labelsize=18)
            ax.grid(True)

        # Add a single legend below the plots
        fig.legend(
            handles=handles, 
            labels=labels, 
            loc='lower center', 
            ncol=4, 
            bbox_to_anchor=(0.5, -0.1), 
            title="Model-Strategy",
            fontsize=24,  # Font size for the legend text
            title_fontsize=24,  # Font size for the legend title
            borderaxespad=1.5  # Padding around the legend box
        )

        # # Adjust layout to fit all subplots and the legend
        fig.tight_layout(rect=[0.0, 0.05, 1, 1])  # Reserve space for the legend
        plt.savefig(f"{base_path}/{dataset}.pdf",format='pdf',dpi=200)
        plt.show()

# Example usage
dataset_names = ["coco"]
model_names = ["VILA1.5-3b", "VILA1.5-13b"]
strategy_names = ["random", "rice"]
metric_names = ["accuracy"]
metric_display = ["Accuracy"]
base_path = "/home/asureddy_umass_edu/cs682/metrics_results/keypoint_detection/mcq_style"

plot_metrics_subplots(dataset_names, model_names, strategy_names, metric_names, metric_display, base_path)
```
